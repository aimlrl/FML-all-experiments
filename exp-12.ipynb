```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import MeanShift

# Step 1: Generate a synthetic 2D dataset
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Visualize the generated data
plt.scatter(X[:, 0], X[:, 1], s=30)
plt.title('Generated Synthetic 2D Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Step 2: Apply Mean Shift algorithm
mean_shift = MeanShift()
mean_shift.fit(X)
labels = mean_shift.labels_
cluster_centers = mean_shift.cluster_centers_

# Visualize the clustering result
plt.scatter(X[:, 0], X[:, 1], c=labels, s=30)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], s=200, marker='X', c='red')
plt.title('Mean Shift Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Print the number of clusters found
print(f"Number of clusters found: {len(np.unique(labels))}")
```
